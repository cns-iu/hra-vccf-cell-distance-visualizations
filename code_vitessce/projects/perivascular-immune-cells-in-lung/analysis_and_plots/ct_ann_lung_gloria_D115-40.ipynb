```python
import numpy as np 
import pandas as pd 
```

```python
pd.set_option('display.max_columns', None)
```

```python
# Read channel marker to cell type key
marker_to_ct_key = pd.read_excel("ChannelMarker-to-CellTypeKey.xlsx")
marker_to_ct_key
marker_to_ct_key
```

```python
# Rename AGER/RAGE to AGER RAGE to match the cell_table column names
# Replace the specified value in column 'A'
marker_to_ct_key['Marker'] = marker_to_ct_key['Marker'].replace("AGER/RAGE", "AGER RAGE")
marker_to_ct_key['Marker'] = marker_to_ct_key['Marker'].apply(lambda x: x.lower())
```

```python
marker_to_ct_key
```

```python
# Read cell table
filename_cell_table = "D115-RLL-10A3-40"
cell_table = pd.read_csv(f"{filename_cell_table}.csv", low_memory=False)
cell_table.head(5)
```

```python
# For D115, replace ACTA2 with SMA in Dataframe. They are both synonyms.
# Iterate through all columns
for col in cell_table.columns:
    # Check if the column name contains "ACTA2"
    if "ACTA2" in col:
        print(f"Replaced column: {col}")
        # Replace "ACTA2" with "SMA"
        new_col = col.replace("ACTA2", "SMA")
        # Rename the column in the DataFrame
        cell_table.rename(columns={col: new_col}, inplace=True)

cell_table.columns
```

```python
# List of column names to consider
columns_to_keep = [col for col in cell_table.columns if 'Cell' in col]
columns_to_check = [col for col in columns_to_keep if 'Mean' in col]
columns_to_check
```

```python

print(f"Initial column count: {len(columns_to_check)}")
#For D115
columns_to_remove = ['DAPI: Cell: Mean',
                     'COL1A1: Cell: Mean',
                        'CD298: Cell: Mean',
                        'Collagen IV: Cell: Mean',
                        'Ki67: Cell: Mean',
                        'Pan-Cytokeratin: Cell: Mean',
                        'CD14: Cell: Mean',
                        'CD3e: Cell: Mean',
                        'HLA-DR: Cell: Mean',
                        'CXCL4: Cell: Mean',
                        'CD45: Cell: Mean',
                        'FOXP3: Cell: Mean',
                        'CD11c: Cell: Mean',
                        'CD163: Cell: Mean',
                        'SCEL: Cell: Mean',
                        'CD1c: Cell: Mean',
                        'VIM: Cell: Mean',
                        'Keratin 14: Cell: Mean'
                     ]
#For D265
# columns_to_remove = ['DAPI: Cell: Mean',
#                      'Col1A1: Cell: Mean',
#                         'CD298: Cell: Mean',
#                         'Collagen IV: Cell: Mean',
#                         'Ki67: Cell: Mean',
#                         'Pan-Cytokeratin: Cell: Mean',
#                         'CD14: Cell: Mean',
#                         'CD3e: Cell: Mean',
#                         'HLA-DR: Cell: Mean',
#                         'CXCL4: Cell: Mean',
#                         'CD45: Cell: Mean',
#                         'FOXP3: Cell: Mean',
#                         'CD11c: Cell: Mean',
#                         'CD163: Cell: Mean',
#                         'SCEL: Cell: Mean',
#                         'CD1c: Cell: Mean',
#                      ]

for val in columns_to_remove:
    if val in columns_to_check:
        columns_to_check.remove(val)

# Drop columns that do not contain the term "Mean". Below two lines are unnecessary. 
df_temp = cell_table[columns_to_check]
df_temp.head(5)
```

```python
len(columns_to_check)
```

```python
columns_to_check
```

```python
# Find the column with the maximum value for each row
cell_table['Max_Marker'] = cell_table[columns_to_check].idxmax(axis=1)
```

```python
cell_table.head(5)
```

```python
# Column to process
column_to_process = 'Max_Marker'
# Iterate through all rows in the specified column and remove ": Cell:Mean"
cell_table[column_to_process] = cell_table[column_to_process].apply(lambda x: x.replace(': Cell: Mean', ''))
cell_table[column_to_process] = cell_table[column_to_process].apply(lambda x: x.lower())
```

```python
cell_table.head(5)
```

```python
# Merge the two DataFrames based on the 'Marker' column
merged_df = pd.merge(cell_table, marker_to_ct_key, left_on='Max_Marker', right_on='Marker', how='left')

# # Replace values in 'Max_Marker' with corresponding values from 'Cell'
# merged_df['Max_Marker'] = merged_df['Cell']
```

```python
merged_df.head(5)
```

```python
# Drop unnecessary columns
merged_df.drop(['Mask Color'], axis=1, inplace=True)
```

```python
merged_df.head(5)
```

```python
merged_df["Cell Type"] = merged_df["Cell Type"].apply(lambda x: x.replace('Macrophage', 'CD68+ Macrophage'))
merged_df["Cell Type"] = merged_df["Cell Type"].apply(lambda x: x.replace('Neutrophil', 'MPO+'))
```

```python
# Specify the columns you want to display
selected_columns = ['Name', 'Max_Marker', 'Marker', 'Cell Type']

# Display only the selected columns
df_selected = merged_df[selected_columns]

# Display the updated DataFrame
df_selected.head(20)
```

```python
merged_df.to_csv(f"Annotated-{filename_cell_table}.csv", index=False)
```
